```python
import numpy as np
import numpy.random as random
import scipy as sp

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

%precision 3

from scipy import interpolate
```

# 練習問題 5-10

```python
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.sin(x**2/5.0)
plt.plot(x,y, 'o')
plt.grid(True)
```

```python
f = interpolate.interp1d(x,y,'linear')
plt.plot(x,f(x),'-')
plt.grid(True)
```

# 練習問題 5-11

```python
f2 = interpolate.interp1d(x,y,'quadratic')
xnew = np.linspace(0, 10, num=30, endpoint=True)

plt.plot(x,y,'o',xnew,f(xnew),'-',xnew,f2(xnew), '--')

plt.legend(['data', 'linear', 'quadratic'], loc='best')
plt.grid(True)
```

# 練習問題 5-11

```python
f2 = interpolate.interp1d(x,y,'quadratic')
f3 = interpolate.interp1d(x,y,'cubic')

xnew = np.linspace(0, 10, num=30, endpoint=True)

plt.plot(x,y,'o',xnew,f(xnew),'-',xnew,f2(xnew), '--', xnew, f3(xnew), '--')

plt.legend(['data', 'linear', 'quadratic', 'cybic'], loc='best')
plt.grid(True)
```
