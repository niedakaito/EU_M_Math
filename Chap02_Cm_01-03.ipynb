# 総合問題2-1,2,3

```python
import numpy as np
import matplotlib.pyplot as plt
import math


# Chap02_Cm_01
x = np.random.uniform(0.0, 1.0, 10000)
y = np.random.uniform(0.0, 1.0, 10000)

# print(x)
# print(y)


# Chap02_Cm_02
# 円の描画
circle = plt.Circle((0,0), 1, color='blue', fill = False)

# 正方形の描画
square = plt.Rectangle((-1, -1), 2, 2, color = 'red', fill = False)

plt.gca().add_patch(circle)
plt.gca().add_patch(square)

distances = [math.hypot(x, y) for x, y in zip(x, y)]

count_less_than_1 = np.sum(np.array(distances) < 1)

print("1より小さい個数:", count_less_than_1)
print("1以上の個数:", 10000 - count_less_than_1)
x_less_than_1 = [x for x, d in zip(x, distances) if d < 1]
y_less_than_1 = [y for y, d in zip(y, distances) if d < 1]

x_over_1 = [x for x, d in zip(x, distances) if d >= 1]
y_over_1 = [y for y, d in zip(y, distances) if d >= 1]

plt.scatter(x_less_than_1, y_less_than_1, color='blue', alpha=0.5, marker='.')
plt.scatter(x_over_1, y_over_1, color='red', alpha=0.5, marker='.')

# グラフの設定
plt.ylim(-1.5, 1.5)
plt.xlim(-1.5, 1.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()


# Chap02_Cm_03
pi = (count_less_than_1 * 4) / 10000
print(pi)
```
