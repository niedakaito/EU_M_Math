```python
import numpy as np
import numpy.random as random
import scipy as sp

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

%precision 3

from scipy import interpolate

from scipy import integrate
import math

from scipy.optimize import fsolve
```

# 練習問題 5-17

```python
def f(x):
  y = 5 * x -10
  return y

x = np.linspace(0,4)
plt.plot(x, f(x))
plt.plot(x, np.zeros(len(x)))
plt.grid(True)

x =fsolve(f, 2)
print(x)
```

# 練習問題 5-18

```python
def f2(x):
  y = x**3 - 2*x**2 -11*x + 12
  return y

x = np.linspace(-5,5)
plt.plot(x, f2(x))
plt.plot(x, np.zeros(len(x)))
plt.grid(True)

x =fsolve(f2, -3)
print(x)

x =fsolve(f2, 1)
print(x)

x =fsolve(f2, 4)
print(x)
```
