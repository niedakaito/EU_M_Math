```python
import numpy.random as random
import numpy as np
import matplotlib.pyplot as plt

random.seed(0)
x = np.random.randn(30)
y = np.sin(x) + np.random.randn(30)

plt.figure(figsize=(20, 6))
plt.plot(x,y,'o')

plt.title('Title Name')
plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)
```
